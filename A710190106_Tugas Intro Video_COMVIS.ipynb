```python
## Taruh kodemu di bawah ini. Beri komentar selengkap lengkapnya
# ...
gedung_edu = cv2.imread("gedung_edu.jpg", cv2.IMREAD_COLOR)
plt.imshow(gedung_edu[:,:,::-1])
print(gedung_edu.shape)
```

```python
layer_pti_blank = np.zeros(gedung_edu.shape, dtype=np.uint8)
```

```python
#resize gambar pti yellow
pti = cv2.imread("pti_yellow.png", cv2.IMREAD_COLOR)
pti_res = cv2.resize(pti ,None,fx=0.4, fy=0.4)
print(pti_res.shape)
```

```python
## Menampilkan bukti konten untuk bingkai

# Tampilan tulisan pada pojok kiri atas di posisi y = 100 dan x = 10
top_left_tulisan = np.array([100, 10], dtype=np.uint16)


# Menempel tulisan pada layer. Sekitar kiri tengah
layer_pti = copy.deepcopy(layer_pti_blank)
layer_pti[top_left_tulisan[0]:top_left_tulisan[0]+pti_res.shape[0], top_left_tulisan[1]:top_left_tulisan[1]+pti_res.shape[1]  ] = pti_res
g = cv2.addWeighted(gedung_edu,1, layer_pti, 0.6 , 0)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(layer_pti[:,:,-1], cmap='gray');
plt.subplot(122);plt.imshow(g[:,:,::-1]);
```

# New Section

```python
#inisiasi codec video, codec untuk tugas disini adalah mp4v
fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
fps = 30 #mengatur fps menjadi 30 frame per detik
lebar = gedung_edu.shape[1] 
tinggi = gedung_edu.shape[0]

video = cv2.VideoWriter('videoFitri.mp4', fourcc, fps, (lebar,tinggi)) #membuat video menggunakan cv2

for i in range(200): #perulangan untuk memindahkan posisi tulisan

    top_left_tulisan = np.array([100+i , 10], dtype=np.uint16) #mengatur tulisan ke posisi bawah, ditunjukan dengan penambahan iterasi pada sumbu Y
    #menggabungkan layer
    layer_pti = copy.deepcopy(layer_pti_blank)
    layer_pti[top_left_tulisan[0]:top_left_tulisan[0]+pti_res.shape[0], top_left_tulisan[1]:top_left_tulisan[1]+pti_res.shape[1]  ] = pti_res
    g = cv2.addWeighted(gedung_edu, 1, layer_pti, 0.6 , 0)
    #menambahkan teks
    text = "Frame: "+ str(i) + "  FPS: 30"
    cv2.putText(g,text, (20, 20), cv2.FONT_HERSHEY_PLAIN, 1.3, (255,255,0), 2, cv2.LINE_AA);
    text2 = "NIM: A710190106, Nama: FITRI WULAN SARI"
    cv2.putText(g,text2, (20, 40), cv2.FONT_HERSHEY_PLAIN, 1.3, (255, 255, 0), 2, cv2.LINE_AA);
    video.write(g)
```
